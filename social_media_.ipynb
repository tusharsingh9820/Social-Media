##Converting the column such as followers,Eng. (Auth.),Eng. (Avg.) in Instagram.csv & Followers,\nViews (Avg.),Likes (Avg.),Comments (Avg.)	in Youtube.csv into integer in order to be used in tableau.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
insta = pd.read_csv('/content/INSTAGRAM.csv', encoding='latin-1')
insta
```

```python
yt = pd.read_csv('/content/YOUTUBE.csv', encoding='latin-1')
yt
```

```python
insta.rename(columns={'Eng. (Auth.)':'authentic views','Eng. (Avg.)':'average views'},inplace = True)
yt.rename(columns={'\nViews (Avg.)':'avg_views'},inplace = True)
```

```python
yt
```

###**Converting column into integer**

```python
def convert_to_int(x):
    if 'M' in x:
        return int(float(x[:-1]) * 1000000)
    if 'K' in x:
        return int(float(x[:-1]) * 1000)
    else:
        return int(float(x))

insta['followers'] = insta['followers'].apply(convert_to_int)
insta['authentic views'] = insta['authentic views'].apply(convert_to_int)
insta['average views'] = insta['average views'].apply(convert_to_int)
yt['Followers'] = yt['Followers'].apply(convert_to_int)
yt['Likes (Avg.)'] = yt['Likes (Avg.)'].apply(convert_to_int)
yt['Comments (Avg.)'] = yt['Comments (Avg.)'].apply(convert_to_int)
yt['avg_views'] = yt['avg_views'].apply(convert_to_int)
```
